난이도: ★★★★☆  
문제3: 위에서 추출된 [DayDate]를 통해 각 요일별 인기 영상들의 [categoryId]는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라  
[사용필드: DayDate, categoryId]  
[사용함수: groupby, as_index, pivot, 통계함수(size)]

```python
# 풀이법
# 1. 대상이 되는 필드 그룹화 하기
# 2. 피벗 테이블을 사용하여 적용하기
```

```python
# DayDate와 categoryId를 기준으로 그룹화하고, 각 그룹의 크기를 계산합니다.
grouped = Youtube.groupby(['DayDate', 'categoryId']).size().reset_index(name='count')

# 피벗 테이블을 생성하여 요일별로 categoryId별 동영상 수를 표시합니다.
pivot_table = grouped.pivot(index='DayDate', columns='categoryId', values='count').fillna(0)

# 인덱스를 다시 설정하여 필요한 형태의 데이터 프레임을 얻습니다.
pivot_table.reset_index(inplace=True)
pivot_table
```

```python
# 가이드 / 힌트
# 0. 일시를 datetime 타입으로 변경하기
# 1. 월 추출하기
# 2. 월마다 계절 필드 생성하기 (map 함수를 사용할 것)
# 3. 계절 그룹화 하기
# ※ 필드 정리하기
```

```python
Seoul.head(1)
```

```python
def find_season(x):
    xx = x // 3
    if xx == 1:
        return '봄'
    elif xx == 2:
        return '여름'
    elif xx == 3:
        return '가을'
    else:
        return '겨울'
```

```python
Seoul['월'] = pd.to_datetime(Seoul['일시']).dt.month
Seoul['계절'] = Seoul['월'].apply(find_season)
Seoul.head(5)
```

```python
Seoul.groupby('계절')[['기온', '풍속', '습도', '현지기압', '해면기압']].agg('mean')
```
