# 문제 1. 주어진 데이터 시각화
임의의 난수로 생성한 두 데이터 집합 $A_1$, $A_2$이 있습니다. 각 데이터 집합의 범위는 아래와 같이 주어집니다.  
$$A_1 = \left\{\left(x_1,y_1\right)|-1\le x_1\le 0, -1\le y_1\le 0\right\}$$  
$$A_2 = \left\{\left(x_2,y_2\right)|2\le x_2\le 4, 2\le y_2\le 4\right\}$$  
주어진 범위에 네 개의 넘파이 배열 $x_1$, $x_2$, $y_1$, $y_2$를 생성하고 하나의 figure에 들어오게 시각화하세요. 시각화할 때의 제약조건은 아래와 같습니다.  
1. 네 개의 넘파이 배열 $x_1$, $x_2$, $y_1$, $y_2$는 공통적으로 20개의 난수로 구성됩니다.</br>(Hint: `numpy.random.rand()` 함수를  활용하세요.)
2. Figure의 사이즈는 `(7, 7)`으로 설정합니다.
3. Figure에 출력하는 x축과 y축의 범위는 각각 $-1\le x\le 4$, $-1\le y\le 4$로 제한시킵니다.
4. 각 포인트를 산점도(scatterplot)으로 그리세요. 이때 `plot()` 메서드를 활용해도 좋고, `scatter()` 메서드를 활용해도 좋습니다. 단, 두 집합 $A_1$, $A_2$를 시각화할 때 두 집합을 표현하는 마커의 종류와 색이 서로 다른 것이어야 합니다. 예를 들어 $A_1$이 빨간색 원형 마커라면, $A_2$는 파란색 삼각형 마커로 표현하거나 하면 됩니다.
5. 범례(legend)를 표시하되, 그 위치는 figure 좌측 상단에 배치합니다. 범례의 내용은 두 집합의 이름인 $A_1$, $A_2$로 넣습니다.
6. 격자(grid)를 표시해주세요.
7. x축과 y축에 라벨을 붙이세요. 단, 라벨의 내용과 폰트 크기 등은 원하는 대로 적어도 무방합니다.

```python
num = 20
x1 = np.random.uniform(-1,0,num)
x2 = np.random.uniform(2,4,num)
y1 = np.random.uniform(-1,0,num)
y2 = np.random.uniform(2,4,num)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x1, y1, color='red', label='class1', marker='o')
plt.scatter(x2, y2, color='blue', label='class2', marker='x')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
np.random.seed(1)
res1 = np.random.randn(1000) * 0.2 + 0.4
res2 = np.random.randn(1000) * 0.2 + 0.4
```

```python
res3 = np.random.uniform(1, 10, 5)
res4 = np.random.uniform(1, 10, 5) + 5

y_ = np.linspace(1, 10, 5)
```

```python
y = np.linspace(-0.5,1.5, 1000)
```

```python
plt.figure(figsize=(5,4))
plt.scatter(res1, y, alpha=0.4)
plt.scatter(res2, y, alpha=0.4)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.hist(res1, histtype='step', bins=20, label='class1')
plt.hist(res2, histtype='step', bins=20, label='class2')
plt.legend()
plt.show()
```
